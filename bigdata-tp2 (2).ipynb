```python
import pandas as pd
import time
import psutil

file_path_gz = "/kaggle/working/emar_detail.csv.gz"

chunk_size = 50_000  # Ø¹Ø¯Ø¯ Ø§Ù„ØµÙÙˆÙ ÙÙŠ ÙƒÙ„ Ø¬Ø²Ø¡
chunks = []


start_time = time.time()
start_memory = psutil.Process().memory_info().rss / (1024 ** 2)  # Ø¨Ø§Ù„Ù…ÙŠØºØ§Ø¨Ø§ÙŠØª

for i, chunk in enumerate(pd.read_csv(file_path_gz, chunksize=chunk_size)):
    print(f"ğŸ“¦ ØªÙ…Øª Ù‚Ø±Ø§Ø¡Ø© Ø§Ù„Ø¬Ø²Ø¡ Ø±Ù‚Ù… {i+1}ØŒ Ø´ÙƒÙ„Ù‡ {chunk.shape}")
    
    
    means = chunk.select_dtypes(include=['number']).mean()
    chunks.append(means)


global_mean = pd.concat(chunks, axis=1).mean(axis=1)


end_time = time.time()
end_memory = psutil.Process().memory_info().rss / (1024 ** 2)

execution_time_compression = end_time - start_time
used_memory_compression = end_memory - start_memory


print("\nâœ… Ù…ØªÙˆØ³Ø·Ø§Øª Ø§Ù„Ø£Ø¹Ù…Ø¯Ø© Ø§Ù„Ø±Ù‚Ù…ÙŠØ© ÙÙŠ ÙƒÙ„ Ø§Ù„Ù…Ù„Ù:")
print(global_mean)

print(f"\nâ±ï¸ ÙˆÙ‚Øª Ø§Ù„ØªÙ†ÙÙŠØ° Ø§Ù„ÙƒÙ„ÙŠ: {execution_time_compression:.2f} Ø«Ø§Ù†ÙŠØ©")
print(f"ğŸ’¾ Ø§Ù„Ø°Ø§ÙƒØ±Ø© Ø§Ù„Ù…Ø³ØªØ®Ø¯Ù…Ø© Ø£Ø«Ù†Ø§Ø¡ Ø§Ù„ØªÙ†ÙÙŠØ°: {used_memory_compression:.2f} Ù…ÙŠØºØ§Ø¨Ø§ÙŠØª")
```
